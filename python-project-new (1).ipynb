```python
# Library Management System

library = {
    "books": [],
    "issued": {}
}

def add_book():
    book = input("Enter book name to add: ")
    if book:
        library["books"].append(book)
        print(f"'{book}' added to library.\n")
    else:
        print("Book name cannot be empty!\n")

def remove_book():
    book = input("Enter book name to remove: ")
    if book in library["books"]:
        library["books"].remove(book)
        print(f"'{book}' removed from library.\n")
    else:
        print("Book not found in library!\n")

def issue_book():
    book = input("Enter book name to issue: ").strip()
    if book in library["books"]:
        user = input("Enter your name: ").strip()
        library["books"].remove(book)
        library["issued"][book] = user
        print(f"'{book}' has been issued to {user}.\n")
    else:
        print("Book is not available!\n")

def return_book():
    book = input("Enter book name to return: ")
    if book in library["issued"]:
        library["books"].append(book)
        user = library["issued"].pop(book)
        print(f"'{book}' returned by {user}.\n")
    else:
        print("This book was not issued.\n")

def view_books():
    print("\n Available Books:")
    if library["books"]:
        for i, book in enumerate(library["books"], start=1):
            print(f"{i}. {book}")
    else:
        print("No books available.\n")

def view_issued_books():
    print("\n Issued Books:")
    if library["issued"]:
        for book, user in library["issued"].items():
            print(f"'{book}' issued to {user}")
    else:
        print("No books are currently issued.\n")

def update_book():
    book = input("Enter the book name you want to update: ")

    # Update from available books
    if book in library["books"]:
        new_name = input("Enter the new book name: ")
        if new_name:
            index = library["books"].index(book)
            library["books"][index] = new_name
            print(f"Book name updated from '{book}' to '{new_name}'.\n")
        else:
            print("New name cannot be empty!\n")

    # Update from issued books
    elif book in library["issued"]:
        new_name = input("Enter the new book name: ")
        if new_name:
            user = library["issued"].pop(book)
            library["issued"][new_name] = user
            print(f"Issued book name updated from '{book}' to '{new_name}'.\n")
        else:
            print("New name cannot be empty!\n")

    else:
        print("Book not found in library or issued list.\n")

def menu():
    while True:
        print("\n--- Library Management System ---")
        print("1. Add Book")
        print("2. Remove Book")
        print("3. Issue Book")
        print("4. Return Book")
        print("5. View All Books")
        print("6. View Issued Books")
        print("7. Update Book Name")
        print("8. Exit")
        choice = input("Enter your choice (1-8): ")

        if choice == '1':
            add_book()
        elif choice == '2':
            remove_book()
        elif choice == '3':
            issue_book()
        elif choice == '4':
            return_book()
        elif choice == '5':
            view_books()
        elif choice == '6':
            view_issued_books()
        elif choice == '7':
            update_book()
        elif choice == '8':
            print("Exiting Library System. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.\n")

# Run the program
menu()
```
